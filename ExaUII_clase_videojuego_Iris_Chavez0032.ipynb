#### (Crear la clase "Videojuego", con 3 atributos (nombre, genero, plataforma), 2 funciones leerDatos(), mostrarDatos() y 2 objetos (juego1, juego2) )

![](https://raw.githubusercontent.com/a25308051280032-boop/ExaUII-Clase-Videojuego-Iris-Chavez-0032/refs/heads/main/infoclasevideojuego.png)

```python
class Videojuego:
    def __init__(self, nombre, genero, plataforma):
        self.nombre = nombre
        self.genero = genero
        self.plataforma = plataforma

    def leerDatos(self):
        self.nombre = input("Ingrese el nombre del videojuego: ")
        self.genero = input("Ingrese el género (ej. RPG, Acción): ")
        self.plataforma = input("Ingrese la plataforma (ej. PC, PS5, Switch): ")

    def mostrarDatos(self):
        print("Videojuego:", self.nombre)
        print("Género:", self.genero)
        print("Plataforma:", self.plataforma)

# Creando el objeto juego1
print("--- Registro de Videojuego 1 ---")
juego1 = Videojuego("", "", "")
juego1.leerDatos()
juego1.mostrarDatos()
print("--------------------------------")

# Creando el objeto juego2
print("--- Registro de Videojuego 2 ---")
juego2 = Videojuego("", "", "")
juego2.leerDatos()
juego2.mostrarDatos()
print("--------------------------------")

print("Software realizado por Iris Chavez NL = 11.")
```
